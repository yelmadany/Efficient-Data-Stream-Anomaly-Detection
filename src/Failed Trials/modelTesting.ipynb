```python
#libraries
import data
import numpy as np
from pyod.models.mad import MAD
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
#Train Data Creation
trainingGenerator = data.generateCPUData(contamination = 10, trainMode=True)
Training_Data = []

for i in range(35065):
  Training_Data.append(next(trainingGenerator))
```

```python
Training_Data
```

```python
data = pd.Series(Training_Data, index=pd.date_range(start = "1-1-2024", end = "1-1-2028", freq="1h"))
```

```python
data
```

```python
from statsmodels.tsa.seasonal import STL

stl = STL(data, seasonal=12)
res = stl.fit()
fig = res.plot()

fig.show()
```

```python
clf = MAD(contamination=0.1)
clf.fit(Training_Data)
```

```python
for i in range(100):
  datapoint = np.array(data.generateCPUData()).reshape(-1, 1)
  is_outlier = clf.predict(datapoint)
  score = clf.decision_function(datapoint)

  print('Outlier:', is_outlier)
  print('Score', score)
```
