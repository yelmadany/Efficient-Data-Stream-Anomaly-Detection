```python
#libraries
import matplotlib.pyplot as plt
import data
import numpy as np
from pyod.models.mad import MAD
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
from statsmodels.tsa.seasonal import STL

%matplotlib inline
#Train Data Creation
trainingGenerator = data.generateCPUData(contamination = 10, trainMode=True)
Training_Data = []

for i in range(35065):
  Training_Data.append(next(trainingGenerator))

data = pd.Series(Training_Data, index=pd.date_range(start = "1-1-2024", end = "1-1-2028", freq="1h"))


stl = STL(data, seasonal=13)
res = stl.fit()
fig = res.plot()

fig.show()
```
