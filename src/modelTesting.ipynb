```python
#libraries
import data
import numpy as np
from pyod.models.mad import MAD
```

```python
#Train Data Creation
Training_Data = []

for i in range(10000):
  Training_Data.append(data.generateCPUData(contamination = 0))

Training_Data = np.array(Training_Data).reshape(-1, 1)
```

```python
Training_Data
```

```python
clf = MAD(contamination=0.1)
clf.fit(Training_Data)
```

```python
for i in range(100):
  datapoint = np.array(data.generateCPUData()).reshape(-1, 1)
  is_outlier = clf.predict(datapoint)
  score = clf.decision_function(datapoint)

  print('Outlier:', is_outlier)
  print('Score', score)
```
